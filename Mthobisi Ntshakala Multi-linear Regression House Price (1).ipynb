# House Price Prediction Using Multi Linear Regression 

```python
import pandas as pd
house_df = pd.read_excel(r"Downloads\\house_price_dataset.xlsx")
house_df.head(3)
```

### changing location into Rural = 1, Suburban = 2 and Urban = 3

```python
house_df['Location'] = house_df['Location'].map({'Rural':1, 'Suburban':2, 'Urban':3})
```

```python
house_df.isnull().sum()
```

```python
house_df.info()
```

```python
house_df.head(2)
```

### using the Location , Size_sqft, Num_Bedrooms, Num_Bathrooms, and Has_Garage to Build a Muilt-linear regression to predict a house price

```python
# Selecting dependent and independent variebles
x = house_df[['Location', 'Size_sqft', 'Num_Bedrooms', 'Num_Bathrooms', 'Has_Garden']]
y = house_df['House_Price']
```

```python
# Spliting the data into Training and tesing Dataset 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.20)
```

```python
#fiting my model using Training Data
from sklearn.linear_model import LinearRegression
house_price_model = LinearRegression()
house_price_model.fit(x_train, y_train)
```

```python
#Predicting the house_price using the Testing Data
house_price_predicted =house_price_model.predict(x_test)
house_price_predicted
```

```python
# Evaluation the model using r2_score using the Testing data
from sklearn.metrics import r2_score 
house_price_r2_score = r2_score( house_price_predicted, y_test)
house_price_r2_score
```
